```python
class Bag:
    def __init__(self, init_data=[]):
        self.data = init_data
        self.len = len(self.data)

    def add(self, x):
        self.data.append(x)
        self.len = len(self.data)

    def addtwice(self, x):
        self.add(x)
        self.add(x)
```

```python
instance = Bag(['a', 'b', 'c'])
instance.data, instance.len
instance.add(1)
instance.add('two')
instance.data, instance.len
# instance.addtwice(3)
# instance.data
```

```python
# parent class
class Person():
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def display(self):
    print(self.name, self.age)

# child class
class Student(Person):
  def __init__(self, name, age):
    self.sName = name
    self.sAge = age
    # inheriting the properties of parent class
    super().__init__("Rahul", age)

  def displayInfo(self):
    print(self.sName, self.sAge)
```

```python
student = Student('Student', 20)
student.name, student.age , student.sName, student.sAge
```

```python
import torch.nn as nn
import torch.nn.functional as F

class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 20, 5)
        self.conv2 = nn.Conv2d(20, 20, 5)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        return F.relu(self.conv2(x))
```

```python
import torch
cnn_model = CNN()
out = cnn_model.forward(torch.rand(10, 1, 28, 28))
out.shape
```

```python
import numpy as np

class CustomDataLoader:
  def __init__(self, inputs, targets):
    self.indexes = list(range(len(inputs)))

  def shuffle(self):
      random.shuffle(self.indexes)

  def get_iter(self, batch_size = 1):
      for start in range(0, len(self.indexes), batch_size):
        if start + batch_size <= len(self.indexes):
          # X_b = [X[id] for id in self.indexes[start:start+batch_size]]
          # y_b = [y[id] for id in self.indexes[start:start+batch_size]]

          X_b  = X[np.array(self.indexes[start:start+batch_size])]
          y_b = y[np.array(self.indexes[start:start+batch_size])]

          yield X_b, y_b, self.indexes[start:start+batch_size]

```

```python
class CustomDataLoaderNP:
  def __init__(self, inputs, targets):
    self.indexes = np.arange(len(inputs))

  def shuffle(self):
      np.random.shuffle(self.indexes)

  def get_iter(self, batch_size = 1):
      for start in range(0, len(self.indexes), batch_size):
        if start + batch_size <= len(self.indexes):

          X_b  = X[self.indexes[start:start+batch_size]]
          y_b = y[self.indexes[start:start+batch_size]]

          yield X_b, y_b, self.indexes[start:start+batch_size]
```

```python
## Implement a dataloader
## Shuffle, divide data into batches,

X, y = load_iris(return_X_y=True)
X.shape, y.shape

data_loader = CustomDataLoaderNP(X, y)

data_loader.shuffle()
data_loader.indexes

counter = 0
for X_b, y_b, id_b in data_loader.get_iter(7):
  print (X_b.shape, y_b.shape)
  # print (id_b)
  counter += 1

counter
```
