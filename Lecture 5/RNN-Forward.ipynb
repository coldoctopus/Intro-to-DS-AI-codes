```python
import torch
import torch.nn as nn

# Input parameters
L = 100  # Sequence length
batch_size = 1
Hin = 5  # Input size

# Create random input tensor
input = torch.randn(L, batch_size, Hin)

# Define the RNN layer
rnn_layer = nn.RNN(input_size=Hin, hidden_size=128)
output, h = rnn_layer(input)

# Check the shape of the RNN output
print(f"RNN output shape: {output.shape}")

# Define the Linear layer
linear_layer = nn.Linear(in_features=128, out_features=10)

# Apply the Linear layer to each time step output from the RNN
# Output shape from RNN is (L, batch_size, hidden_size)
# We want to apply the linear layer to the last dimension (hidden_size -> 10)
linear_output = linear_layer(output)

# Check the shape of the final output
print(f"Final output shape: {linear_output.shape}")
```
